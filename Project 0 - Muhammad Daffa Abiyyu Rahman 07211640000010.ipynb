We start with the importing, using the script below :

```python
import cv2 as cv
import numpy as np
import scipy as sc
from matplotlib import pyplot as plt
```

The script above grants the access to all the installed component which will be needed for the next part of the script. After all of the imports are completed, we begin loading the image as shown below :

```python
img = cv.imread("Sample.jpg")
```

This script loads the image that will be used. It's <b>necessary to have the file in the same directory as the script and have the script pointed to the right directory of the image</b>. Now, let's see some informations from the image using numpy functions using the scripts below :

```python
img.shape
```

```python
img.size
```

```python
img.min(),img.max()
```

```python
img.mean()
```

For this image, it has 640x480 size with 3 channels and a pixel size of 921600 with min value of 0 and max value of 255 with mean of 82.24. It's possible to retrieve numerical informations and manipulate images using numpy functions.

Subsequently, we will show the image using the script below :

```python
plt.imshow(img)
```

Above is a sample image used with the result showing a blue dominant image. This is caused by how imshow color channel works, which is an inverse of RGB the image is based. It uses BGR channel. In addition, there are some axis information which might be needed at all. If the axis is to be discarded, the portion of the script needs to be changed into :

```python
plt.axis("OFF")
plt.imshow(img)
```

This way, the image will no longer show the axis and only show the image with BGR channel. This might not be what one desired, and it would be necessary to flip the channel.

After some search, I personally discovered from PyImageSearch website that there's this OpenCV command that can be used to reverse the RGB channel. The function is called cvtColor() and the constant for flipping the channel is called cv2.COLOR_BGR2RGB which now will be used to yield the image in RGB channel :

```python
plt.axis("OFF")
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
```

With the cvtColor() function, now the image is properly shown as how it originally is. There are other channel options in cvtColor(), but for now this is sufficient for the aim of the assignment.

## Final Script
Below is the final script for loading an image and then showing the image with the proper RGB format. print() is added to properly show both the geometry and the pixel size of the image in the result information. show() is used to ensure the image is properly shown in a window in any mode (non-interactive or interactive) and avoid showing the memory information in some interactive mode (like the output examples above).

```python
import cv2 as cv
import numpy as np
import scipy as sc
from matplotlib import pyplot as plt

img = cv.imread("Sample.jpg")

print(img.shape)
print(img.size)
print(img.min(),img.max())

plt.axis("OFF")
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()
```
